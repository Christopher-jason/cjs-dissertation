```python
import pandas as pd
import yfinance
import matplotlib.pyplot as plt
import numpy as np

#get the data
s_p = pd.read_csv('price_data/top_13/S&P.csv')
nvda = pd.read_csv('price_data/top_13/NVDA.csv')
tsla = pd.read_csv('price_data/top_13/TSLA.csv')
ubs = pd.read_csv('price_data/top_13/UBS.csv')
```

```python
#plot closing price of s_p from march 2023 to august 2023
plt.plot(s_p[s_p['Date'] ], s_p['Close'])
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('S&P 500')
plt.show()

```

```python
ubs['Date'] = pd.to_datetime(s_p['Date'])
start_date = pd.to_datetime('2023-03-16')
end_date = pd.to_datetime('2023-08-04')

filtered_data = ubs[(ubs['Date'] >= start_date) & (ubs['Date'] <= end_date)]
```

```python
plt.figure(figsize=(9, 6))
plt.plot(filtered_data['Date'], filtered_data['Close'], label='Closing Price', color='b')
plt.title('Closing Prices - UBS')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()


plt.show()
```
